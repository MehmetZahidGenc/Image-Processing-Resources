```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('tesla.png', 0)

plt.title('Original Image')
plt.imshow(image, cmap='gray')
```

```python
def bit_plane_slicing(image, bit_planes=[6, 7]):
  bit_planes_image = np.full_like(image, 0)

  for bit_plane in bit_planes:
    bit_planes_image  += np.bitwise_and(image, np.full_like(image, 2**bit_plane))

  return bit_planes_image
```

```python
# Display all bits one by one

outputs = []

for i in range(8):
  outputs.append(bit_plane_slicing(image, bit_planes=[i]))


plt.rcParams["figure.figsize"] = [10, 10]
plt.rcParams["figure.autolayout"] = True

fig, ax = plt.subplots(nrows=2, ncols=4)

# Row 1
ax[0][0].axis('off')
ax[0][0].title.set_text('Bit: 0')
ax[0][0].imshow(outputs[0], cmap='gray')

ax[0][1].axis('off')
ax[0][1].title.set_text('Bit: 1')
ax[0][1].imshow(outputs[1], cmap='gray')

ax[0][2].axis('off')
ax[0][2].title.set_text('Bit: 2')
ax[0][2].imshow(outputs[2], cmap='gray')

ax[0][3].axis('off')
ax[0][3].title.set_text('Bit: 3')
ax[0][3].imshow(outputs[3], cmap='gray')

# Row 2
ax[1][0].axis('off')
ax[1][0].title.set_text('Bit: 4')
ax[1][0].imshow(outputs[4], cmap='gray')

ax[1][1].axis('off')
ax[1][1].title.set_text('Bit: 5')
ax[1][1].imshow(outputs[5], cmap='gray')

ax[1][2].axis('off')
ax[1][2].title.set_text('Bit: 6')
ax[1][2].imshow(outputs[6], cmap='gray')

ax[1][3].axis('off')
ax[1][3].title.set_text('Bit: 7')
ax[1][3].imshow(outputs[7], cmap='gray')
```

```python
"""
As we can see in the results, the last bits is that contain the most information.

Now, let's get a picture using only the last 3 bits
"""

output_image = bit_plane_slicing(image, bit_planes=[5, 6, 7])

# Display images
plt.rcParams["figure.figsize"] = [10, 10]
plt.rcParams["figure.autolayout"] = True

fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].axis('off')
ax[0].title.set_text('Original Image')
ax[0].imshow(image, cmap='gray')

ax[1].axis('off')
ax[1].title.set_text('the image formed from the last 3 bits')
ax[1].imshow(output_image, cmap='gray')
```

```python
"""
As we saw above, the last three bits contained the most information. 
We created a new image from these bits and compared it with the original image.
The result image is almost equivalent to the original image.

However,  when we show the bits one by one, we see that the fourth bit also contains good information. 
Now let's add also fourth bit and look at our results.

"""
output_image = bit_plane_slicing(image, bit_planes=[4, 5, 6, 7])

# Display images
plt.rcParams["figure.figsize"] = [10, 10]
plt.rcParams["figure.autolayout"] = True

fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].axis('off')
ax[0].title.set_text('Original Image')
ax[0].imshow(image, cmap='gray')

ax[1].axis('off')
ax[1].title.set_text('the image formed from the last 4 bits')
ax[1].imshow(output_image, cmap='gray')
```
