```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

"""
There are 3 types of gradient filters (or called as High-pass filter); Sobel, Scharr and Laplacian
"""

# Original
image = cv2.imread('gradients.png', 0)

# Laplacian
laplacian = cv2.Laplacian(image, cv2.CV_64F)

# Sobel-x
sobel_x = cv2.Sobel(image, cv2.CV_64F,1,0,ksize=5)

# Sobel-y
sobel_y = cv2.Sobel(image, cv2.CV_64F,0,1,ksize=5)

# Display images
plt.rcParams["figure.figsize"] = [7, 7]
plt.rcParams["figure.autolayout"] = True

fig, ax = plt.subplots(nrows=2, ncols=2)


# Row 1
ax[0][0].axis('off')
ax[0][0].title.set_text('Original')
ax[0][0].imshow(image, cmap='gray')

ax[0][1].axis('off')
ax[0][1].title.set_text('Laplacian')
ax[0][1].imshow(laplacian, cmap='gray')

# Row 2
ax[1][0].axis('off')
ax[1][0].title.set_text('Sobel-x')
ax[1][0].imshow(sobel_x, cmap='gray')

ax[1][1].axis('off')
ax[1][1].title.set_text('Sobel-y')
ax[1][1].imshow(sobel_y, cmap='gray')
```
