```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

"""
There are 3 types of gradient filters (or called as High-pass filter); Sobel, Scharr and Laplacian
"""

# Original
image = cv2.imread('daisy.jpg', 0)

# cv2.Canny(source_img, min_val_threshold, max_val_threshold)

# 1
edges_1 = cv2.Canny(image, 100, 200)

# 2
edges_2 = cv2.Canny(image, 50, 120)

# 3
edges_3 = cv2.Canny(image, 150, 210)

# Display images
plt.rcParams["figure.figsize"] = [7, 7]
plt.rcParams["figure.autolayout"] = True

fig, ax = plt.subplots(nrows=2, ncols=2)


# Row 1
ax[0][0].axis('off')
ax[0][0].title.set_text('Original')
ax[0][0].imshow(image, cmap='gray')

ax[0][1].axis('off')
ax[0][1].title.set_text('edges_1')
ax[0][1].imshow(edges_1, cmap='gray')

# Row 2
ax[1][0].axis('off')
ax[1][0].title.set_text('edges_2')
ax[1][0].imshow(edges_2, cmap='gray')

ax[1][1].axis('off')
ax[1][1].title.set_text('edges_3')
ax[1][1].imshow(edges_3, cmap='gray')
```
