```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('daisy.jpg')

# We read the image with opencv so our image is in BGR color space.
# There are lots of type color spaces. You can use which is more useful for you.
img_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

result_image = np.hstack((image, img_hsv, img_rgb)) # all images display side by side

plt.figure(figsize=(10, 10))
plt.title('Result Image')
plt.imshow(result_image, cmap='gray')
```
