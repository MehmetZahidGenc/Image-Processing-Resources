```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

# Original
img = cv2.imread('morph.png')

kernel = np.ones((6,6),np.uint8)

# Erosion
erosion = cv2.erode(img,kernel,iterations = 1)

# Dilation
dilation = cv2.dilate(img,kernel,iterations = 1)

# Morphological Gradient
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

# Top hat
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)

# Black Hat
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)


# Display images
plt.rcParams["figure.figsize"] = [7, 7]
plt.rcParams["figure.autolayout"] = True

fig, ax = plt.subplots(nrows=2, ncols=3)

# Row 1
ax[0][0].axis('off')
ax[0][0].title.set_text('original')
ax[0][0].imshow(img, cmap='gray')

ax[0][1].axis('off')
ax[0][1].title.set_text('Erosion')
ax[0][1].imshow(erosion, cmap='gray')

ax[0][2].axis('off')
ax[0][2].title.set_text('Dilation')
ax[0][2].imshow(dilation, cmap='gray')

# Row 2
ax[1][0].axis('off')
ax[1][0].title.set_text('Morphological Gradient')
ax[1][0].imshow(gradient, cmap='gray')

ax[1][1].axis('off')
ax[1][1].title.set_text('Top hat')
ax[1][1].imshow(tophat, cmap='gray')

ax[1][2].axis('off')
ax[1][2].title.set_text('Black Hat')
ax[1][2].imshow(blackhat, cmap='gray')
```

```python
# Opening
img_open = cv2.imread('opening.png')
opening = cv2.morphologyEx(img_open, cv2.MORPH_OPEN, kernel)

# Closing
img_close = cv2.imread('closing.png')
closing = cv2.morphologyEx(img_close, cv2.MORPH_CLOSE, kernel)

# Display images
plt.rcParams["figure.figsize"] = [7, 7]
plt.rcParams["figure.autolayout"] = True

fig, ax = plt.subplots(nrows=2, ncols=2)


# Row 1
ax[0][0].axis('off')
ax[0][0].title.set_text('img_open')
ax[0][0].imshow(img_open, cmap='gray')

ax[0][1].axis('off')
ax[0][1].title.set_text('opening')
ax[0][1].imshow(opening, cmap='gray')

# Row 2
ax[1][0].axis('off')
ax[1][0].title.set_text('img_close')
ax[1][0].imshow(img_close, cmap='gray')

ax[1][1].axis('off')
ax[1][1].title.set_text('closing')
ax[1][1].imshow(closing, cmap='gray')
```
