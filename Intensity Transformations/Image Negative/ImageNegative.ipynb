```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from scipy.signal import correlate2d
```

```python
image = cv2.imread('mamogram.png', 0) # our image has one channel

plt.rcParams["figure.figsize"] = [10, 10]
plt.title('original mamogram image')
plt.imshow(image, cmap='gray')
plt.show()
```

```python
"""
Negative of Image

-> Although it is mostly used in medical imaging systems, 
   it is used in many different fields, space technologies are the main ones.

-> It is used to enrich the white or gray details hidden in dark or light areas in the image

"""
# input image must be one-channel image
def neg_of_image(image):
  L = np.max(image)
  img_negative = L - image

  return img_negative

img_negative = neg_of_image(image=image)

# Display images
plt.rcParams["figure.figsize"] = [10, 10]
plt.rcParams["figure.autolayout"] = True

fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].axis('off')
ax[0].title.set_text('original')
ax[0].imshow(image, cmap='gray')

ax[1].axis('off')
ax[1].title.set_text('Negative of Image')
ax[1].imshow(img_negative, cmap='gray')
```
